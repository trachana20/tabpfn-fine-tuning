```python
import arff
import csv

def arff_to_csv(arff_file, csv_file):
    with open(arff_file, 'r') as f:
        data = arff.load(f)
    
    with open(csv_file, 'w', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(data['attributes'])
        writer.writerows(data['data'])
```

```python
arff_file = '/Users/rachana/Downloads/phpMYEkMl.arff'  # Replace with your ARFF file path
csv_file = '/Users/rachana/Downloads/titanic.csv'   # Replace with your desired CSV file path

arff_to_csv(arff_file, csv_file)
print(f'ARFF file "{arff_file}" successfully converted to CSV file "{csv_file}"')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns  # Optional for enhanced visuals

def visualize_feature_distributions(csv_file):
    # Load the dataset into a Pandas DataFrame
    df = pd.read_csv(csv_file)
    
    # Select numeric columns 
    numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
    
    # Plot distributions for each numeric feature
    for col in numeric_columns:
        plt.figure(figsize=(8, 5))
        
        # Example of different types of plots you can create
        sns.histplot(df[col], kde=True)  # Histogram with kernel density estimate
        # sns.kdeplot(df[col], shade=True)  # Kernel Density Estimate plot
        # sns.boxplot(x=df[col])  # Box plot
        
        plt.title(f'Distribution of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')
        plt.grid(True)
        plt.tight_layout()
        plt.show()

    
visualize_feature_distributions(csv_file)
```
